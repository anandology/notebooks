# Parametric Equations and Rose Curves

The common approach to plot graphs involves plotting a function $y=f(x)$. This works with some shapes, but  many interesting curves can't be drawn in this approach. 

The $y = f(x)$ form of equations are called *rectangular equations*, they have a limitation that there is only one value of y for every value of x. 

With that we can plot simple lines and curves like parabola with a vertical axis of symmetry, but can not draw a cicle or a parabola with a horizontal axis of symmetry.

## Parametric Equations

Parameteric equations formutaed as:
    
$x = f(t) \\ y=g(t)$

They use a special parameter $t$ to compute both $x$ and $y$

For example, we could use the following parameteric equations to draw a parabola with horizontal axis of symmetry.

$x = t^2 \\ y = t$

We could use $t$ in range $[-1, 1]$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-1, 1, 1000)
x = t*t
y = t

plt.figure(figsize=(4, 4))
plt.plot(x, y)
plt.grid()
```

Similarly, we could plot a circle using the following parametric equations.

$x = cos(t) \\ y = sin(t)$ 

for $t$ in $[0, 2\pi]$.

```python
t = np.linspace(0, 2*np.pi, 1000)
x = np.cos(t)
y = np.sin(t)

plt.figure(figsize=(4, 4))
plt.plot(x, y)
plt.grid()
```

Another interesting example of a parametric plot is [Lassajous Curve][1]. They are defined as:

$x = sin(at+\delta) \\ y=sin(bt)$


[1]: https://en.wikipedia.org/wiki/Lissajous_curve

```python
delta = np.pi/2
a = 3
b = 4

t = np.linspace(0, 2*np.pi, 1000)
x = np.sin(a*t + delta)
y = np.sin(b*t)

plt.figure(figsize=(4, 4))
plt.plot(x, y)
plt.grid()
```

Code to generate the images used above.

**Rectangular Equations**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(12, 4))

x = np.linspace(-1, 1, 1000)
y0 = 2*x+3
y1 = x*x
y2 = x*x*x - 0.8*x*x 

ax0.plot(x, y0)
ax0.grid()
ax0.set_title("Stright line")

ax1.plot(x, y1)
ax1.grid()
ax1.set_title("Parabola with vertical axis of symmetry")

ax2.plot(x, y2)
ax2.grid()
ax2.set_title("Cubic curve");

plt.tight_layout()

plt.savefig("rectangular.svg")
```

**Parametric equations**

```python
fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(12, 4))

t = np.linspace(0, 2*np.pi)

x0 = np.cos(t)
y0 = np.sin(t)

ax0.plot(x0, y0)
ax0.grid()
ax0.set_title("Circle")

y1 = x*x
y2 = x*x*x - 0.8*x*x 


t1 = np.linspace(-1, 1, 100)
x1 = t1*t1
y1 = t1
ax1.plot(x1, y1)
ax1.grid()
ax1.set_title("Parabola with horizontal axis of symmetry")

t2 = np.linspace(0, 2*np.pi, 1000)
x2 = np.cos(t2)**3
y2 = np.sin(t2)**3

ax2.plot(x2, y2)
ax2.grid()
ax2.set_title("Astroid");

plt.tight_layout()

plt.savefig("parametric.svg")
```
